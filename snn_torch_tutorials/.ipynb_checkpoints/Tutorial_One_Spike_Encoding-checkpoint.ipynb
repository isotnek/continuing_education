# Tutorial on Spike Encoding with snnTorch

Original source and additional information can be found [here](https://snntorch.readthedocs.io/en/latest/tutorials/tutorial_1.html)

```python
import torch
from torch.utils.data import DataLoader

import snntorch as snn
from snntorch import utils, spikegen

from torchvision import datasets, transforms
```

```python
# Training Parameters

batch_size = 128
data_path = './data/mnist'
num_classes = 10

#Torch variables
dtype = torch.float
```

```python
# Download dataset and define transform

transform = transforms.Compose([
    transforms.Resize((28,28)),
    transforms.Grayscale(),
    transforms.ToTensor(),
    transforms.Normalize((0,), (1,))
])

mnist_train = datasets.MNIST(data_path, train = True, download = True, transform = transform)
```

```python
# Create a subset of the data for easier manipulation

subset = 10
mnist_train = utils.data_subset(mnist_train, subset)
```

```python
# Create dataloader

train_loader = DataLoader(mnist_train, batch_size = batch_size, shuffle = True)
```

```python
# Temporal Dynamics

num_steps = 100

# Create vector of elements of value = 0.5
raw_vector = torch.ones(num_steps) * 0.5

# Pass each sample through Bernoulli Trial
rate_coded_vector = torch.bernoulli(raw_vector)

print(f'The output is spiking {rate_coded_vector.sum() * 100 / len(rate_coded_vector):.2f}% of the time')
```

```python
# Iterate through minibatches

data = iter(train_loader)
data_it, targets_it = next(data)

spike_data = spikegen.rate(data_it, num_steps = num_steps)

print(spike_data.size())
```

# Visualizing Simulated Spiking Neurons

```python
import matplotlib.pyplot as plt
import snntorch.spikeplot as splt
from IPython.display import HTML
```

```python
spike_data_sample = spike_data[:, 0, 0]
print(spike_data_sample.size())
```

```python
fig, ax = plt.subplots()
anim = splt.animator(spike_data_sample, fig, ax)
plt.rcParams['animation.ffmpeg_path'] = '/opt/homebrew/bin/ffmpeg'

HTML(anim.to_html5_video())
```

```python
print(f'The corresponding target is: {targets_it[0]}')
```

```python
# Reducing the spiking rate by manipulating the gain

spike_data = spikegen.rate(data_it, num_steps = num_steps, gain = 0.25)

spike_data_sample2 = spike_data[:, 0, 0]
fig, ax = plt.subplots()
anim = splt.animator(spike_data_sample2, fig, ax)
plt.rcParams['animation.ffmpeg_path'] = '/opt/homebrew/bin/ffmpeg'

HTML(anim.to_html5_video())
```

```python
# Averaging spikes over time & reconstructing input images
plt.figure(facecolor = 'w')
plt.subplot(1, 2, 1)
plt.imshow(spike_data_sample.mean(axis = 0).reshape((28, -1)).cpu(), cmap = 'binary')
#plt.axis('off')
plt.title('Gain = 1')

plt.subplot(1, 2, 2)
plt.imshow(spike_data_sample2.mean(axis = 0).reshape((28, -1)).cpu(), cmap = 'binary')
#plt.axis = ('off')
plt.title('Gain = 0.25')

plt.show()
```

## Raster Plot

```python
spike_data_sample2 = spike_data_sample2.reshape((num_steps, -1))

fig = plt.figure(facecolor = 'w', figsize = (10,5))
ax = fig.add_subplot(111)
splt.raster(spike_data_sample2, ax, s = 1.5, c = 'black')

plt.title('Input Layer')
plt.xlabel('Time Step')
plt.ylabel('Neuron Number')
plt.show()
```

```python
# Index into a single neuron

idx = 600 #index the nth neuron

fig = plt.figure(facecolor = 'w', figsize = (8, 1))
ax = fig.add_subplot(111)

splt.raster(spike_data_sample.reshape(num_steps, -1)[:, idx].unsqueeze(1), ax, s = 10, c = 'black', marker = '|')

plt.title('Input Neuron')
plt.xlabel('Time Step')
plt.yticks([])

plt.show()
```

## Latency Coding

```python
def convert_to_time(data, tau = 5, threshold = 0.01):
    spike_time = tau * torch.log(data / (data - threshold))
    return spike_time
```

```python
raw_input = torch.arange(0, 5, 0.05)
spike_times = convert_to_time(raw_input)
```

```python
plt.plot(raw_input, spike_times)
plt.xlabel('Input Value')
plt.ylabel('Spike Time (s)')
plt.show()
```
