```python
# importing relevant libraries

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#set seaborn style
sns.set(style = 'white', context = 'notebook', rc = {'figure.figsize' : (14, 10)})
```

```python
mnist = load_digits()
mnist.keys()
```

```python
X = pd.DataFrame(mnist.data)
X.head()
```

```python
y = pd.DataFrame(mnist.target)
y.head()
```

```python
#standardize data
std_X = StandardScaler().fit_transform(X)
print(std_X.shape)
```

```python
#applying PCA
pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(std_X)

principalDF = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])

principalDF.shape
```

```python
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
#plotting the first 2 principal components
plt.scatter(principalComponents[:, 0],
            principalComponents[:, 1],
            s = 5,
            c = y,
            cmap = 'Spectral'
)
plt.gca().set_aspect('equal', 'datalim')
plt.colorbar(boundaries = np.arange(11) - 0.5).set_ticks(np.arange(10))
plt.title('Principal Components of Sklearn Digits', fontsize = 24);
plt.xlabel('PC 1')
plt.ylabel('PC 2')
```

```python
#trying again with 3 principal components
pca = PCA(n_components = 3)
principalComponents = pca.fit_transform(X)

principalDF = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2', 'PC3'])

principalDF.shape
```

```python
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
#figuring out how many principal components to use
pca = PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```
